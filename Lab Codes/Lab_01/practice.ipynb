```python
# Answer to the question no 1
# Implement the tanh activation function where tanH(x) = (e^x - e^-x) / (e^x + e^-x)
import numpy as np
studentId = 129
z = studentId % 5

tanH = (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

print(tanH)
```
